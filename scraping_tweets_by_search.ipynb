# Scraping de tweets através de busca

### Passo 1: encontrando pessoas com depressão no twitter
Para identificar pessoas com depressão, buscaremos tweets que contenham:
- "diagnóstico de depressão" mas que não contenham sem diagnóstico"

```python
FILENAME = 'tweets'
TERMS = [
    '"diagnósticado com depressão"  -"sem diagnóstico"',
    '"diagnósticada com depressão"  -"sem diagnóstico"'
]
```

```python
import twint
from datetime import datetime, timedelta
import nest_asyncio
nest_asyncio.apply()
```

```python
# Método recebe um palavra chave e salva em um csv os resultados encotrados
def search_tweets(keyword):
    c = twint.Config()
    c.Search = keyword
    c.Custom['tweet'] = ['username','tweet', 'link']
#     c.Since = datetime.strftime(datetime.now() - timedelta(1), '%Y-%m-%d')
#     c.Until = datetime.strftime(datetime.now(), '%Y-%m-%d')
    c.Store_csv= True
    c.Store_object = True
    c.Output = FILENAME+'.csv'
    
    twint.run.Search(c)
```

```python
def search_by_terms():
    for term in TERMS:
        search_tweets(term)
```

```python
search_by_terms()
```
